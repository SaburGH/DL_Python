```python

import noisereduce as nr
import librosa
import soundfile as sf

# Load the audio file
y, sr = librosa.load("/content/drive/MyDrive/Colab Notebooks/white-noise-soft-333297.mp3", sr=None)
```

```python
from IPython.display import Audio
Audio('/content/drive/MyDrive/Colab Notebooks/white-noise-soft-333297.mp3')
```

```python
# Estimate noise using the first 0.5 seconds of the audio
noise_sample = y[0:int(sr*0.5)]

# Reduce noise
reduced_noise = nr.reduce_noise(y=y, sr=sr, y_noise=noise_sample)

# Save the result
sf.write("cleaned_audio.wav", reduced_noise, sr)
```
