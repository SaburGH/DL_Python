## Filtering

```python
import numpy as np
import matplotlib.pyplot as plt

import obspy


# Read the seismogram
#st = obspy.read("https://examples.obspy.org/RJOB_061005_072159.ehz.new")
#st='1001.segd'
# There is only one trace in the Stream object, let's work on that trace...
#tr = st[0]

#tr=df[0][100]

jam=20
tr=df[0][jam]
# Filtering with a lowpass on a copy of the original Trace
tr_filt = tr.copy()
tr_filt.filter('lowpass', freq=60.0, corners=2, zerophase=True)

# Now let's plot the raw and filtered data...
t = np.arange(0, tr.stats.npts / tr.stats.sampling_rate, tr.stats.delta)
plt.subplot(211)
plt.title(f'Trace {jam} from SEGD file: {segd_file_path} ')
plt.plot(t, tr.data, 'k')
plt.ylabel('Raw Data')
plt.subplot(212)
plt.plot(t, tr_filt.data, 'k')
plt.ylabel('Lowpassed Data')
plt.xlabel('Time [s]')
plt.suptitle(tr.stats.starttime)
plt.show()



jam1=80
tr=df[0][jam1]
# Filtering with a lowpass on a copy of the original Trace
tr_filt = tr.copy()
tr_filt.filter('lowpass', freq=60.0, corners=2, zerophase=True)

# Now let's plot the raw and filtered data...
t = np.arange(0, tr.stats.npts / tr.stats.sampling_rate, tr.stats.delta)
plt.subplot(211)
plt.title(f'Trace {jam1} from SEGD file: {segd_file_path} ')
plt.plot(t, tr.data, 'k')
plt.ylabel('Raw Data')
plt.subplot(212)
plt.plot(t, tr_filt.data, 'k')
plt.ylabel('Lowpassed Data')
plt.xlabel('Time [s]')
plt.suptitle(tr.stats.starttime)
plt.show()


jam2=110
tr=df[0][jam2]
# Filtering with a lowpass on a copy of the original Trace
tr_filt = tr.copy()
tr_filt.filter('lowpass', freq=60.0, corners=2, zerophase=True)

# Now let's plot the raw and filtered data...
t = np.arange(0, tr.stats.npts / tr.stats.sampling_rate, tr.stats.delta)
plt.subplot(211)
plt.title(f'Trace {jam2} from SEGD file: {segd_file_path} ')
plt.plot(t, tr.data, 'k')
plt.ylabel('Raw Data')
plt.subplot(212)
#plt.title(f'Trace {jam} from SEGD file: {segd_file_path} ')
plt.plot(t, tr_filt.data, 'k')
plt.ylabel('Lowpassed Data')
plt.xlabel('Time [s]')
plt.suptitle(tr.stats.starttime)
plt.show()
```

## Example

```python
import numpy as np
import matplotlib.pyplot as plt

 # Generating a sample musical note signal
fs = 150  # Sampling frequency (Hz)
duration = 1  # seconds
frequency = 8  # A4 note frequency (Hz)
```

```python
t = np.linspace(0, duration, int(fs * duration), endpoint=False)
sigs=np.sin(2 * np.pi * frequency * t)
sigs
# Plotting the spectrum
plt.plot(sigs)
plt.title('FFT of a Musical Note')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
t = np.linspace(0, duration, int(fs * duration), endpoint=False)
sigc=np.cos(2 * np.pi * frequency * t)
sigc
# Plotting the spectrum
plt.plot(sigc)
plt.title('FFT of a Musical Note')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
plt.plot(sigs)
plt.plot(sigc)
```

```python
noise = np.random.normal(0, 1, len(t))  # Signal with noise
# Plotting the spectrum
plt.plot(noise)
plt.title('FFT of a Musical Note')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
```

```python
s = sigs + np.random.normal(0, 1, len(t))  # Signal with noise
# Plotting the spectrum
plt.plot(s)
plt.title('FFT of a Musical Note')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
```

```python
# Applying FFT
#fft_result = np.fft.fft(signal)
fft_result = np.fft.fft(s)
freq = np.fft.fftfreq(t.shape[-1], d=1/fs)

# Plotting the spectrum
plt.plot(freq, np.abs(fft_result))
plt.title('FFT of a Musical Note')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
```

## Frequecy vs Time

```python
import numpy as np

from obspy.signal.tf_misfit import plot_tfr


# general constants
tmax = 6.
dt = 0.01
npts = int(tmax / dt + 1)
t = np.linspace(0., tmax, npts)

fmin = .5
fmax = 10

# constants for the signal
A1 = 4.
t1 = 2.
f1 = 2.
phi1 = 0.

# generate the signal
H1 = (np.sign(t - t1) + 1) / 2
st1 = A1 * (t - t1) * np.exp(-2 * (t - t1))
st1 *= np.cos(2. * np.pi * f1 * (t - t1) + phi1 * np.pi) * H1

plot_tfr(st1, dt=dt, fmin=fmin, fmax=fmax)
```
