```python
#img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/image_processing_batch-2/images/orange-1.jpg', cv2.IMREAD_GRAYSCALE)


img = cv2.imread('orange-1.jpg', cv2.IMREAD_GRAYSCALE)
```

### Global Thresholding

```python
plt.imshow(img, cmap='gray')
```

```python
# ret, thresholded = cv.threshold(img, 230, 255, cv.THRESH_BINARY)

ret, thresholded = cv2.threshold(img, 230, 255, cv2.THRESH_BINARY)
plt.imshow(thresholded, cmap='gray')
print(ret)
```

```python
ret, thresholded = cv2.threshold(img, 230, 255, cv2.THRESH_BINARY_INV)
plt.imshow(thresholded, cmap='gray')
print(ret)
```

```python
ret, thresholded = cv2.threshold(img, 230, 0, cv2.THRESH_BINARY_INV)
plt.imshow(thresholded, cmap='gray')
print(ret)
```

```python
ret, thresholded = cv2.threshold(img, 250, 255, cv2.THRESH_TRUNC)
plt.imshow(thresholded, cmap='gray')
print(ret)
```

```python
ret, thresholded = cv2.threshold(img, 150, 255, cv2.THRESH_TOZERO)
plt.imshow(thresholded, cmap='gray')
print(ret)
```
