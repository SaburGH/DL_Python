### Rotation

```python
# Rotate 90 degrees clockwise
rotated_image = cv2.rotate(image, cv2.ROTATE_180)

# Display the original and rotated images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(rotated_image)
plt.title('Rotated Image')

plt.show()
```


use the cv2.getRotationMatrix2D() function to create a rotation matrix and then apply it to the image using cv2.warpAffine()

```python
image.shape
```

```python
# Get the image dimensions
(h, w) = image.shape[:2]

# Calculate the center of the image
center = (w // 2, h // 2)

# Create a rotation matrix
M1 = cv2.getRotationMatrix2D(center, 30, 1.0)
M2 = cv2.getRotationMatrix2D(center, 45, 1.0)
M3 = cv2.getRotationMatrix2D(center, -90, 1)


# Rotate the image
rotated_image_30 = cv2.warpAffine(image, M1, (w, h))
rotated_image_45 = cv2.warpAffine(image, M2, (w, h))
rotated_image_60 = cv2.warpAffine(image, M3, (w, h))


plt.figure(figsize=(15,5))

plt.subplot(131)
plt.imshow(rotated_image_30)
plt.title('30 degree rotated')

plt.subplot(132)
plt.imshow(rotated_image_45)
plt.title('45 degree rotated')

plt.subplot(133)
plt.imshow(rotated_image_60)
plt.title('60 degree clockwise rotated')

plt.tight_layout()
plt.show()
```
